```python
import pandas as pd
```

```python
wine = pd.read_excel('F:\Skillenable\Data frames\wine_clustering.xlsx')
wine.head()
```

```python
wine.info()
wine.isnull().sum()   
```

```python
wine.describe().T
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
# Building clustering model
from sklearn.cluster import KMeans
```

```python
sns.scatterplot(wine['Alcohol'], wine['Hue'])
```

```python
# setting range of cluster upto 10
for i in range(1,11):
    print(i)
```

```python
# prepare req data
error= []
for j in range (1,11):
    km = KMeans(n_clusters=j)
    km.fit(wine)        # feed data
    error.append(km.inertia_)   # calculate error
error
```

```python
# plot elbow diagram
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.lineplot (range(1,11), error, marker ='o')
plt.title('Elbow diagram')
plt.show
```

```python
# Select 5 cluster points cause elbow curve tends to falt after 5 points
# Build model with 5 cluster points
km_5 = KMeans(n_clusters= 5).fit(wine)
clusters = km_5.fit_predict(wine)
```

```python
# creating new column 
wine['cluster'] = clusters
wine.head()
```

```python
# identify how many values in each cluster
wine['cluster'].value_counts()
```

```python
# identify individual cluster
wine.groupby('cluster')[['Alcohol','Malic_Acid', 'Ash', 'Ash_Alcanity', 'Magnesium', 'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols', 'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280', 'Proline']].mean()
```

```python
# plot a scatter plot for the the same we just done

sns.scatterplot(wine['Alcohol'], wine['Hue'], hue= wine['cluster'], palette='plasma')
```
